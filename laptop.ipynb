```python
import pandas as pd
import numpy as np
laptop = pd.read_csv(r'./assets/data/laptop-price.csv')
```

Menampilkan Overview Data Laptop Price


```python
laptop.head()
```

Data Cleansing
untuk memastikan data sudah baik dan terstruktur

```python
# Mereplace value Not Available dengan null
laptop['processor_gnrtn'].replace('Not Available', np.nan, inplace=True)
```

```python
# menghilangkan th dibelakang gen procie
laptop['processor_gnrtn'] =laptop['processor_gnrtn'].str.rstrip("th")
# menghilangkan satuan GB
laptop['ram_gb'] =laptop['ram_gb'].str.rstrip("GB")
# menghilangkan satuan GB
laptop['ssd'] =laptop['ssd'].str.rstrip("GB")
# menghilangkan satuan GB
laptop['hdd'] =laptop['hdd'].str.rstrip("GB")
# menghilangkan satuan GB
laptop['graphic_card_gb'] =laptop['graphic_card_gb'].str.rstrip("GB")
# Mereplace value no warranty dengan default value => 0 year
laptop['warranty'] = laptop['warranty'].str.replace("No warranty", "0 year")

# melihat hasil excecute diatas
laptop
```

Replacing


```python
# Mereplace value No dan Yes, menjadi value boolean
laptop['Touchscreen'] = laptop['Touchscreen'].replace({'No': "False", "Yes": "True"})
laptop['msoffice'] = laptop['msoffice'].replace({'No': "False", "Yes": "True"})

# melihat hasil excecute diatas
laptop
```

Formatting

```python
# Merubah Pricing, data ini dari Scrapper India, maka kita convert rupe ke Rupiah, 180.
def convert_to_rupiah(price):
    price_in_rupiah = price * 180
    return f'{price_in_rupiah}'

laptop['Price'] = laptop['Price'].apply(convert_to_rupiah)
laptop
```

```python
# Menghilangkan Satuan Stars di kolom Rating
laptop['rating'] = laptop['rating'].str.rstrip("stars")
laptop
```

```python
# Mengecek Type Data, semuanya object, perlu kita sesuaikan untuk analisa yang lebih dalam
laptop.dtypes
```

```python
# Sebenarnya, banyak data processor_gnrtn yang null, kita perlu isi dengan nilai mean
mean_laptop_procie_gen = round(laptop['processor_gnrtn'].astype(float).mean())
print(mean_laptop_procie_gen)
laptop['processor_gnrtn'].replace(np.nan, mean_laptop_procie_gen, inplace=True)
```

```python
# Setelah checking tipe data, ada beberapa kolom yang perlu kita convert ke numeric
int_columns = ['ram_gb', 'ssd','hdd', 'graphic_card_gb', 'Price', 'Number of Ratings', 'Number of Reviews',]
laptop[int_columns] = laptop[int_columns].apply(pd.to_numeric)
```

```python
laptop['processor_gnrtn'] = pd.to_numeric(laptop['processor_gnrtn'])
```

```python
# Cek counts value tertentu
laptop['warranty'].value_counts()
```

```python
# Menghilangkan satuan year
laptop['warranty'] = laptop['warranty'].str.rstrip("year")
laptop['warranty']
```

```python
# jangan lupa, bukan hanya year, tapi juga ada years
laptop['warranty'] = laptop['warranty'].str.rstrip("years")
```

```python
# convert ke numeric untuk eda lebih lanjut
laptop['rating'] = pd.to_numeric(laptop['rating'])
laptop['warranty'] = pd.to_numeric(laptop['warranty'])
```

```python
# di kolom brand, semuanya capitalize, kecuali acer, maka kita perlu perbaikan data
laptop['brand'].replace("acer", "Acer", inplace=True)
```

```python
# pengecekan tipe data, sudah mulai rapi
laptop.dtypes
```

```python
# standar penamaan features adalah lowercase, dengan separator _
laptop.rename(columns={'Touchscreen': "touchscreen", "Price": "price", "Number of Ratings": "no_ratings", "Number of Reviews": "no_reviews"}, inplace=True)
```

```python
# export ke excel dan csv untuk eda histogram, boxplot, dan statistic descriptive
laptop.to_excel('./assets/clean/laptop-price.xlsx', index=None, header=True)
laptop.to_csv('./assets/clean/laptop-price.csv', index=False)
```
