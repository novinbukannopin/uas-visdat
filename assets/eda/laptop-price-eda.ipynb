COUNTS LAPTOP BY BRAND

```python
ax = laptop['brand'].value_counts().plot(kind="bar", colormap="plasma")
plt.bar_label(ax.containers[0])
plt.ylabel("Counts")
plt.xlabel("Brand")
plt.title("Counts Laptop by Brand")
```

```python
ax = sns.barplot(x=laptop["brand"],y=laptop["rating"], hue=laptop["ram_gb"])
```

```python
sns.barplot(x=laptop["brand"],y=laptop["price"], hue=laptop["ram_gb"])
plt.xticks(rotation="vertical")
plt.ylabel("Price")
plt.xlabel("Brand")
plt.title("Price comparison is influenced by ram")
plt.show()
```

```python
# kepadatan;
ax = sns.histplot(x=laptop["ram_type"], hue=laptop['processor_brand'], bins=10)
ax.set_title("Comparison ram type by brand")
ax.set_xlabel("Ram Type")
```
